**Skewness and Kurtosis**

```python
# Step 3.1: Draw the histogram plot of Total_cost,Load_factor, and Output to gain general insights about the data.
panel_df['Total_cost'].plot(kind="hist") # positive skew
```

```python
panel_df['Load_factor'].plot(kind="hist") # negative skew
```

```python
panel_df['Output'].plot(kind="hist") # positive skew
```

```python
# Step 3.2: calculate skewness and kurtosis of Total_cost,Load_factor, and Output.
print("Total Cost Skew: ",panel_df['Total_cost'].skew())
print("Total Cost Kurtosis:",panel_df['Total_cost'].kurtosis())
```

```python
print("Load Factor Skew: ",panel_df['Load_factor'].skew())
print("Load Factor Kurtosis:",panel_df['Load_factor'].kurtosis())
```

```python
print("Output Skew: ",panel_df['Output'].skew())
print("Output Kurtosis:",panel_df['Output'].kurtosis())
```
