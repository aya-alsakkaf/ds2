```python
# Step 3.2: Select the name and Unit of the products with a price less than or equal to 10.

sql_statement = "SELECT ProductName, Price FROM Products where Price <= 10"
df_price_name_less_than_10 = pd.read_sql_query(sql_statement, conn)
df_price_name_less_than_10
```

```python
# Step 3.3: Select all the products with a price between 5 and 15.

sql_statement = "SELECT ProductName, Price FROM Products where Price between 5 and 15"
df_price_name_between_10_15 = pd.read_sql_query(sql_statement, conn)
df_price_name_between_10_15
```

```python
# Step 3.4: Compute the average price of all the products.

sql_statement = "SELECT AVG(Price) as Average FROM Products";
df_avg_price = pd.read_sql_query(sql_statement, conn)

df_avg_price
```

```python
# Step 3.5: Compute the count of all the products where SupplierID equal to 3.

sql_statement = "SELECT COUNT(*) as Count FROM Products where SupplierID=3"
df_supplierId_3 = pd.read_sql_query(sql_statement, conn)
df_supplierId_3
```

```python
# Step 3.6: Select the name and price of all products with a price larger than or equal to 20, and sort first by name (in descending order), and then by price (in ascending order).

sql_statement = "SELECT ProductName, Price FROM Products where price >= 20 ORDER BY ProductName DESC, Price ASC"
df_sorted = pd.read_sql_query(sql_statement, conn)
df_sorted
```
